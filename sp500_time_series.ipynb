```r
# Load necessary libraries
library(ggplot2)
library(readr)

# Read CSV file
df <- read_csv("data/^SPX.csv")

# Convert 'Date' to Date type
df$Date <- as.Date(df$Date)

# Create the plot
p <- ggplot(df, aes(x = Date, y = `Adj Close`)) +
  geom_point(color = "blue") +
  ggtitle("S&P500 Index Adjusted Close Price Over Time") +
  xlab("Date") +
  ylab("Close Price") +
  theme_minimal() +
  theme(
    plot.background = element_rect(fill = "transparent", color = NA),
    panel.background = element_rect(fill = "transparent", color = NA),
    axis.text.x = element_text(angle = 45, hjust = 1)
  )

# Show the plot
print(p)

# Save with transparent background
ggsave("sp500_plot.png", plot = p, bg = "transparent", width = 10, height = 5, dpi = 300)
```
